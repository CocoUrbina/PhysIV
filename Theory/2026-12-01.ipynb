# OUTLINE

## Maxwell's Equations - Differencial Form

$$ \mathbf{\nabla} \cdot \mathbf{E} = \frac{\rho_E}{\epsilon_0}$$

$$ \mathbf{\nabla} \cdot \mathbf{B} = 0$$

$$ \mathbf{\nabla} \times \mathbf{E} = -\frac{\partial \mathbf{B}}{\partial t}$$

$$ \mathbf{\nabla} \times \mathbf{B} = \mu_0 \epsilon_0 \frac{\partial \mathbf{E}}{\partial t} + \mu_0 \mathbf{J_E}$$

## Maxwell's Equations - Integral Form

$$\int \mathbf{E} \cdot d\mathbf{A} = \frac{Q_E}{\epsilon_0}$$

$$\int \mathbf{B} \cdot d\mathbf{A} = 0$$

$$\int \mathbf{E} \cdot d\mathbf{l} = - \frac{d \Phi_B }{dt}$$

$$\int \mathbf{B} \cdot d\mathbf{d} = \mu_0 \epsilon_0 \frac{d \Phi_E }{dt} + \mu_0 I_E$$
