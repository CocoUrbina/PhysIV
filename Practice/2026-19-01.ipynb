# 2

$$\mathbf{E} = E_0 \cos(\omega t - kz) \mathbf{x}$$ e incide sucesivamente sobre:

1) Polarizador lineal cuyo eje forma un ángulo $\theta$ con respecto al eje $\mathbf{x}$.

a) Determine el campo eléctrico emergente.

b) ¿Para qué valor de $\theta$ el campo es circularmente polarizado?

c) Calcule la intensidad transmitida en función de $\theta$.
