<h3>Pregunta 1</h3>

<p><strong>De las siguientes, ¿cuál tiene una longitud de onda menor si todas tienen la misma energía cinética no relativista?</strong></p>

<form>
  <input type="radio" id="a" name="pregunta1" value="electron">
  <label for="a">a) Electrón</label><br><br>

  <input type="radio" id="b" name="pregunta1" value="proton" >
  <label for="b">b) Protón</label><br><br>

  <input type="radio" id="c" name="pregunta1" value="neutron">
  <label for="c">c) Neutrón</label><br><br>

  <input type="radio" id="d" name="pregunta1" value="sodio" checked>
  <label for="d">d) Átomo de sodio</label>
</form>


**Solución**

De la relación de de Broglie:

$$ \lambda = \frac{h}{p} \tag{1} $$

donde la energía cinética no relativista es

$$ K = \frac{p^2}{2m} \rightarrow p = \sqrt{2mK} \tag{2} $$

Ec. (2) en ec.(1):

$$ \lambda = \frac{h}{\sqrt{2mK}} $$

Para $K$ constante, 

$$ \lambda \propto \frac{1}{\sqrt{m}} $$


Un átomo de sodio está conformado por varios electrones + protones + neutrones, por lo que es el más pesado de entre las 4 opciones. Entonces la opción de menor longitud de onda.

Opción d).

<h3>Pregunta 2</h3>

<p><strong>En las siguientes situaciones, conteste la opción que tiene la partícula (electrón, protón o igual) que tiene una longitud de onda de materia más grande:</strong></p>

<ol>
  <li>Un electrón y un protón con el mismo momento.</li>
  <li>Un electrón y un protón con la misma rapidez.</li>
  <li>Un electrón y un protón con la misma energía cinética no relativista.</li>
</ol>

<form>
  <input type="radio" id="a" name="pregunta2" value="a">
  <label for="a">a) Igual, electrón, igual</label><br>

  <input type="radio" id="b" name="pregunta2" value="b" checked>
  <label for="b">b) Igual, electrón, electrón</label><br>

  <input type="radio" id="c" name="pregunta2" value="c">
  <label for="c">c) Electrón, electrón, igual</label><br>

  <input type="radio" id="d" name="pregunta2" value="d">
  <label for="d">d) Electrón, electrón, electrón</label>
</form>

**Solución**


1. Si tienen el mismo momento, entonces

$$ \lambda_e = \frac{h}{p} = \lambda_p $$

2. Si tienen la misma rapidez, entonces

$$ \lambda_e = \frac{h}{m_e v} $$

$$ \lambda_p = \frac{h}{m_p v} $$

Como $ m_p > m_e $, entonces $ \lambda_e > \lambda_p$.

3. De la pregunta 1, $ \lambda \propto \frac{1}{\sqrt{m}} $

Como $ m_p > m_e $, entonces $ \lambda_e > \lambda_p$.

Opción b).

<h3>Pregunta 3</h3>

<p><strong>La longitud de onda de un electrón que se mueve con una rapidez de $1.63 \times 10^5 ~\mathrm{m/s}$ en nm es:</strong></p>

**Solución**


\begin{align}
\lambda_e &= \frac{h}{m_e v} \\
  &= \frac{6.626\times 10^{-34}}{9.109 \times 10^{-31} \times 1.63 \times 10^5} \\
  &= 4.46\times 10^{-9} ~\mathrm{m}\\
  &= 4.46 ~\mathrm{nm}
\end{align}
