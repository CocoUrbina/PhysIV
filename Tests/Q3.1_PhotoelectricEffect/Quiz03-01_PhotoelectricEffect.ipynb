<h3>Pregunta 1</h3>

<p><strong>
Para cierto metal, electrones son emitidos cuando se irradia con longitudes de onda por debajo de los 435 nm, pero no para longitudes de onda más grandes. Calcule la función de trabajo en eV.</strong></p>

**Solución**
Función de trabajo:

$$ \phi = \frac{hc}{\lambda_0} $$

donde $ \lambda_0 $ es la **longitud de onda umbral**. Entonces,


\begin{align}
\phi &= \frac{hc}{\lambda_0} \\
  &= \frac{6.626\times 10^{-34}~\mathrm{J\cdot s}\times \left(\frac{1~\mathrm{eV}}{1.602\times 10^{-19}~\mathrm{J}}\right) \times 3.0 \times 10^8~\mathrm{m/s}}{435 \times 10^{-9} ~\mathrm{m}} \\
  &= 2.85 ~\mathrm{eV}
\end{align}

<h3>Pregunta 2</h3>

<p><strong>Electrones son emitidos cuando una fuente ultravioleta con longitud de onda $\lambda$ se apunta a un metal. Si queremos duplicar el número de electrones emitidos por minuto, se debería:</strong></p>

  <input type="radio" id="b" name="pregunta1" value="b" >
  <label for="b">b) Reemplazar la fuente por una de longitud de onda 2$\lambda$ </label><br><br>

  <input type="radio" id="c" name="pregunta1" value="c">
  <label for="c">c) Reemplazar la fuente por una del doble de frecuencia</label><br><br>

  <input type="radio" id="d" name="pregunta1" value="d" >
  <label for="d">d) Reemplazar la fuente por una de longitud de onda $\lambda$/2 </label>
</form>

**Solución**


De la ecuación fundamental del efecto fotoeléctrico,

$$ K_{\mathrm{max}} = h\nu - \phi $$

Se observa que:

- la frecuencia $\nu$ afecta la energía, pero no la cantidad de electrones.
- cambiar la longitud de onda $\lambda$ cambia la función de trabajo $\phi$ y por tanto la energía, pero tampoco la cantidad de electrones.

Por otro lado, se observa experimentalmente que

$$ I \propto N_e $$

Donde $N_e$ es la cantidad de electrones emitidos.

Por tanto, duplicar la frecuencia (una segunda fuente), duplica la cantidad de electrones emitidos.

Opción a).

<h3>Pregunta 3</h3>

<p><strong>Considere lo siguiente sobre un experimento sobre efecto fotoeléctrico:</strong></p>

<ol>
  <li>Si se reemplaza la fuente por una del doble de frecuencia, pero que emite el mismo número de fotones por segundo, el potencial de frenado se incrementa.</li>
  <li>Si se reemplaza el material por uno con una función de trabajo menor, el potencial de frenado se disminuye.</li>
</ol>

</form>

**Solución**


De la ecuación fundamental del efecto fotoeléctrico,

$$ K_{\mathrm{max}} = h\nu - \phi \tag{1}$$

El potencial de frenado $V_0$ es, 

$$ eV_0 = K_{\mathrm{max}} \tag{2}$$.


Combinando las ecuaciones (1) y (2),

$$ V_0 = \frac{h\nu -\phi}{e} \tag{3}$$

De ec.(3), se observa que si se aumenta $\nu$, aumenta también $V_0$. **La afirmación 1 es correcta.**

De ec.(3), se observa que si se disminuye $\phi$, aumenta $V_0$. **La afirmación 2 NO es correcta.**

Opción a).
