Considere la siguiente onda electromagnética que se propaga en el vacío:

$$\mathbf{E}(z,t) = (E_0 \hat{i} + i E_0 \hat{j})e^{i(kz - wt)}$$

donde $E_0$, $k$, $w$ son constantes reales positivas e $i = \sqrt{-1}$.

a) Defina qué es una onda. (1 pt)

b) A partir de la Ley de Faraday, derive la componente magnética de esta onda. (4 pts)

c) A partir del ítem anterior o del enunciado, encuentre las componentes **reales** de la onda y determine la polarización de la misma. (5 puntos)

_Recuerde que $e^{i \theta} = \cos \theta + i \sin \theta $ y úselo para mostrar que $i = e^{i\pi / 2}$._ 



**Solución** 

a) Es una perturbación en el tiempo y en el espacio que transfiere energía.

b) 
$$ \mathbf{\nabla} \times \mathbf{E} =  -\frac{\partial \mathbf{B}}{\partial t}$$

$$ \mathbf{\nabla} \times \mathbf{E} = 
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
\partial_x & \partial_y & \partial_z \\
E_x & E_y & E_z
\end{vmatrix} =
\begin{vmatrix}
\mathbf{i} & \mathbf{j} & \mathbf{k} \\
\partial_x & \partial_y & \partial_z \\
E_x & E_y & 0
\end{vmatrix}
= -\mathbf{i} \frac{\partial E_y}{\partial z}
+\mathbf{j} \frac{\partial E_x}{\partial z}
= -\mathbf{i} \frac{\partial \left(i E_0 e^{i(kz - wt)}\right)}{\partial z}
+\mathbf{j} \frac{\partial \left(E_0 e^{i(kz - wt)}\right)}{\partial z}
$$

$$ \mathbf{\nabla} \times \mathbf{E}
= k E_0 e^{i(kz - wt)} \mathbf{i} +
i k E_0 e^{i(kz - wt)} \mathbf{j}
= k E_0 ( \mathbf{i} + i \mathbf{j})  e^{i(kz - wt)}
$$

Es decir, 

$$-\frac{\partial \mathbf{B}}{\partial t} = k E_0 ( \mathbf{i} + i \mathbf{j})  e^{i(kz - wt)} $$

Integrando con respecto a $t$,

\begin{aligned}
\mathbf{B}(z,t)
&=- \int k E_0 ( \mathbf{i} + i \mathbf{j})  e^{i(kz - wt)} \ dt\\
&= - k E_0 ( \mathbf{i} + i \mathbf{j}) \int e^{i(kz - wt)} \ dt\\
&= - k E_0 ( \mathbf{i} + i \mathbf{j}) \frac{e^{i(kz - wt)}}{-i\omega}\\
&=  \frac{k E_0}{i\omega} ( \mathbf{i} + i \mathbf{j}) e^{i(kz - wt)}
\end{aligned}

Como $1/i = -i$ y $k/\omega = 1/c$, 

$$\boxed{\mathbf{B}(z,t) = \frac{E_0}{c} ( -i \mathbf{i} + \mathbf{j}) e^{i(kz - wt)}}$$

c) 

\begin{aligned}
\Re(\mathbf{E})
&= \Re(E_x) + \Re(E_y)\\
&= \Re(E_0 e^{i(kz - wt)}) + \Re(i E_0 e^{i(kz - wt)})\\
&= \Re\{E_0 [\cos(kz - wt) + i \sin (kz - wt)]\} + \Re\{iE_0 [\cos(kz - wt) + i \sin (kz - wt)]\}\\
&= \Re\{E_0 [\cos(kz - wt) + i \sin (kz - wt)]\} + \Re\{E_0 [- \sin (kz - wt)] + i\cos(kz - wt)\}\\
&= E_0 \cos(kz - wt)  - E_0 \sin (kz - wt)\\
&= E_0 \cos(kz - wt) \mathbf{i}  - E_0 \sin (kz - wt) \mathbf{j}
\end{aligned}

Hay un desfase de $\pi/2$ entre ambas componentes. Entonces hay polarización circular.

Determinemos ahora el sentido del giro. Consideremos $z=0$. Entonces, 

\begin{aligned}
\Re(\mathbf{E})
&= \Re(E_x) + \Re(E_y)\\
&= E_0 \cos(- wt) \mathbf{i}  - E_0 \sin (- wt) \mathbf{j}\\
&= E_0 \cos(wt) \mathbf{i}  + E_0 \sin ( wt) \mathbf{j}
\end{aligned}

Conforme $t$ aumenta, $\Re(\mathbf{E})$ va de $\mathbf{i}$ a $\mathbf{j}$. Es decir, **el giro es antihorario**.
