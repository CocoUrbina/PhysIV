  <input type="radio" id="d" name="pregunta1" value="d" >

---
En el marco de Tomás,

$$ \Delta t = 0, \qquad \Delta x \neq 0 $$

Con transformación temporal de Lorentz,

$$ \Delta t^\prime = \gamma \left( \Delta t - \frac{v \Delta x}{c^2} \right) $$

donde $\Delta t = 0$. Entonces queda


$$ \Delta t^\prime = - \gamma \frac{v \Delta x}{c^2} $$

Nótese que $ \Delta t^\prime \neq 0 $ y que el signo depende del signo de $v$.

<h3>Pregunta 3</h3>

<p><strong>Una estrella (en reposo relativo a la Tierra) está a 100 años luz de nuestro planeta. Un astronauta sale rumbo a esta estrella desde la Tierra con una rapidez $v=0.98c$. ¿Cuánto tiempo dura el astronauta en llegar a la estrella según el austronauta?</strong></p>

<form>
  <input type="radio" id="a" name="pregunta1" value="a">
  <label for="a">a) 100 años </label><br><br>

  <input type="radio" id="b" name="pregunta1" value="b">
  <label for="b">b) 98 años </label><br><br>

  <input type="radio" id="c" name="pregunta1" value="c" checked>
  <label for="c">c) 20 años </label><br><br>

  <input type="radio" id="d" name="pregunta1" value="d" >
  <label for="d">d) 102 años </label>
</form>

---

En el sistema del astronauta, la estrella se mueve hacia él y la distancia está contraída por relatividad especial.

\begin{align}
t^\prime &= \frac{d^\prime}{v} \\
 \displaystyle &= \frac{\left(\frac{d}{\gamma}\right)}{v}\\
  &= \frac{d}{\gamma v}\\
   &= \frac{d}{v} \left( 1 - v^2/ c^2 \right)^{1/2}\\
    &= \frac{100c}{0.98c} \left( 1 - 0.98^2 \right)^{1/2}\\
    &\approx 20.3
\end{align}

---

Por conservación del cuadrivector energía-momento:

Se descartan a) y b) porque $ E_0 = mc^2 $ es la energía relativista de una partícula en reposo.

Por otro lado, la conservación de la energía y del momento son válidas si se redefinen como:

$$ E = \gamma mc^2$$

$$ \mathbf{p} = \gamma m \mathbf{v} $$
