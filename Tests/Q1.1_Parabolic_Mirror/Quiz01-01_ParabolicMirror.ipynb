En clase estudiamos que para cualquier espejo cuya forma esté dada por la función $y(x)$ y que sea simémetrica alrededor del eje $y$, su foco está dado por

$$f = y + \frac{x(1-\tan^2 \theta)}{2 \tan \theta} .$$

---

(a) Explique por qué deducíamos que $\tan \theta = dy / dx$ (1 pt)

**Solución**

Porque $dy/dx$ es la pendiente de la recta tangente en ese punto. Si la tangente forma un ángulo $\theta$ con respecto al eje-x,entonces, $\tan \theta = dy / dx$

---

(b) Considere que el espejo es parabólico, cuya ecuación está dada por

$$ y(x) = Ax^2 $$

muestre que $\displaystyle f = \frac{1}{4A}$ e interprete este resultado. (5 pt)

**Solución**

Tenemos que,

$$y(x) = Ax^2$$

Derivando

$$ y^\prime (x) = 2xA $$

como $\tan \theta = dy/dx$

$$ \tan\theta = 2xA \tag{1}$$

El resultado general es

$$f = y + \frac{x(1-\tan^2 \theta)}{2 \tan \theta} \tag{2}$$

De ec.(1).

$$ 1- \tan^2\theta = 1 - 4 x^2 A^2 $$

O bien,

$$ x(1- \tan^2\theta) = x(1 - 4 x^2 A^2) $$

O bien, 

$$ \frac{x(1- \tan^2\theta)}{2\tan\theta} = \frac{x(1 - 4 x^2 A^2)}{2(2xA)} \tag{3}$$

Ec.(3) en ec.(2):


\begin{align}
f &= y +  \frac{x(1 - 4 x^2 A^2)}{2(2xA)}\\
  &= Ax^2 +  \frac{x(1 - 4 x^2 A^2)}{4xA}\\
  &= Ax^2 +  \frac{x}{4xA} - \frac{x4x^2A^2}{4xA}\\
  &= Ax^2 +  \frac{1}{4A} - x^2A\\
  &= \frac{1}{4A}
\end{align}

El resultado anterior quiere decir que el foco no depende de $x$.

(c) La ecuación

$$f = y + \frac{x(1-\tan^2 \theta)}{2 \tan \theta} $$

se puede reescribir como 

$$f = y + \frac{x(1-(dy/dx)^2)}{2 dy/dx} $$

si reemplazamos lo mencionado en el primer ítem. Si se toma la derivada respecto a $x$ del foco, se iguala a 0 y se reacomoda, se obtiene la siguiente ecuación diferencial:

$$ x \frac{d^2y}{dx^2} - \frac{dy}{dx} = 0 $$

Resuelva esta ecuación diferencial y explique cómo su resultado se asocia con el ítem anterior. (4 pt)

**Solución**

$$ x y^{\prime \prime} - y^\prime = 0 $$

Consideremos $ v = y^\prime $. Entonces la EDO queda como

$$ x v^\prime - v = 0 $$

Reacomodando,

$$ x \frac{dv}{dx}  = v  $$

O bien, 

$$ \frac{1}{v} dv  = \frac{1}{x} dx $$

Integrando,

$$ \ln(v) = \ln(x) + C $$

Aplicando la función exponencial,

$$ v = C x  $$

Devolviendo el cambio de variable,

$$ y^\prime = C x  $$

Integrando,

$$ y = C \frac{x^2}{2} + C_2 $$

Que es lo mismo que 

$$ y(x) = C_1 x^2  + C_2 $$.

Se obtiene nuenvamente la ecuación de la parábola.
