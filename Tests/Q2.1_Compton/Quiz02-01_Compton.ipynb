a) Escriba los 4-vectores $(E, \mathbf{p} c)$ para el sistma inicial y final. (3 pt)

b) Usando 4-vectores, muestre que la longitud de onda del fotón dispersado $\lambda^\prime$ es


$$\lambda^\prime = \lambda + \frac{h}{mc} (1 - \cos \theta) $$

**Nota:** recuerde que $$ P^2 = (mc^2)^2 $$. (4 pts)

c) Discuta los siguientes casos límites (3 pt):
- $\theta = 0$
- $\theta = \pi$
- $\theta = \pi$, y $mc^2 << E_\gamma$


**Solución** 

a) 

El 4-momento es $ P = (E; \mathbf{p} c) $

Energía-Momento: $ E^2 = p^2c^2 + m^2 c^4 $, Para todas las partículas. 

<u>Antes de la colisión:</u>

Fotón: $P_\gamma = (E_\gamma; pc, 0, 0)$.

Pero $E_\gamma =pc.$ Entonces:

$P_\gamma = (E_\gamma; E_\gamma , 0, 0)$

Electrón:  $P_e = (mc^2; 0, 0, 0)$.

<u>Después de la colisión:</u>

Fotón: $P_\gamma^\prime = (E_\gamma^\prime; E_\gamma^\prime \cos\theta , E_\gamma^\prime \sin\theta , 0)$

Electrón:  $P_e^\prime = (E_e^\prime; \mathbf{p_e} c)$.


b) 

Por conservación del momento relativista,

\begin{align}
P_i &= P_f \\
P_\gamma + P_e   &=  P_\gamma^\prime + P_e^\prime \\
P_\gamma + P_e - P_\gamma^\prime   &=  P_e^\prime \\
(P_\gamma + P_e - P_\gamma^\prime)^2   &=  (P_e^\prime)^2 \\
P_\gamma^2 + P_e^2 + (P_\gamma^\prime)^2 +2P_\gamma P_e -2P_\gamma^\prime P_\gamma -2P_\gamma^\prime P_e  &=  (P_e^\prime)^2 \\
\end{align}

Veamos ahora los términos cuadrados. Para los 4-vectores $A$ y $B$, 

$\boxed {A B = AB - \mathbf{A}\cdot \mathbf{B}}$

$ P_\gamma^2 = P_\gamma P_\gamma =  (E_\gamma; E_\gamma , 0, 0)(E_\gamma; E_\gamma , 0, 0) = E_\gamma^2 - E_\gamma^2 - 0 -0 = 0 $

$ P_e^2 = P_e P_e =  (mc^2; 0 , 0, 0)(mc^2; 0 , 0, 0) = (mc^2)^2 - 0 - 0 - 0 = (mc^2)^2 $

$ (P_\gamma^\prime)^2 = P_\gamma^\prime P_\gamma^\prime =  (E_\gamma^\prime; E_\gamma^\prime \cos\theta , E_\gamma^\prime \sin\theta , 0)(E_\gamma^\prime; E_\gamma^\prime \cos\theta , E_\gamma^\prime \sin\theta , 0) = (E_\gamma^\prime)^2 - (E_\gamma^\prime)^2 \cos^2 \theta - (E_\gamma^\prime)^2 \sin^2 \theta -0 = 0 $

$ (P_e^\prime)^2 = (P_e^\prime) (P_e^\prime) = (mc^2)^2 $, por la nota que nos dan en el enunciado.


Entonces,

\begin{align}
2P_\gamma^\prime P_\gamma  &=  2P_\gamma P_e - 2P_\gamma^\prime P_e\\
P_\gamma^\prime P_\gamma  &=   P_\gamma P_e - P_\gamma^\prime P_e \\
P_\gamma^\prime P_\gamma  &=    P_e(P_\gamma - P_\gamma^\prime) \\
\end{align}

El lado izquierdo:

$ P_\gamma^\prime P_\gamma = (E_\gamma; E_\gamma, 0, 0)(E_\gamma^\prime; E_\gamma^\prime \cos\theta, E_\gamma^\prime \sin\theta, 0) = E_\gamma E_\gamma^\prime - E_\gamma E_\gamma^\prime \cos\theta = E_\gamma E_\gamma^\prime (1 - \cos\theta)$.

El lado derecho:

$ P_e(P_\gamma - P_\gamma^\prime) = (mc^2;0,0,0) \left((E_\gamma; E_\gamma , 0, 0) - (E_\gamma^\prime; E_\gamma^\prime \cos\theta , E_\gamma^\prime \sin\theta , 0) \right) = (mc^2;0,0,0) (E_\gamma - E_\gamma^\prime; E_\gamma - E_\gamma^\prime \cos\theta, 0 - E_\gamma^\prime \sin\theta, 0-0) = mc^2(E_\gamma - E_\gamma ^\prime) $

Entonces, 

$$ E_\gamma E_\gamma^\prime (1 - \cos\theta) = mc^2(E_\gamma - E_\gamma ^\prime) $$

Considerando que,

$E_\gamma = \frac{hc}{\lambda}$,        $E_\gamma^\prime = \frac{hc}{\lambda^\prime}$

\begin{align}
\frac{hc}{\lambda} \frac{hc}{\lambda^\prime} (1 - \cos\theta) &= mc^2\left(\frac{hc}{\lambda} - \frac{hc}{\lambda^\prime}\right)\\
\frac{h}{mc} (1 - \cos\theta) &= \lambda \lambda^\prime \left( \frac{1}{\lambda} - \frac{1}{\lambda^\prime} \right) \\
\frac{h}{mc} (1 - \cos\theta) &= \lambda^\prime - \lambda \\
\frac{h}{mc} (1 - \cos\theta) +  \lambda &= \lambda^\prime 
\end{align}

(c)

- $\theta = 0$: $\lambda = \lambda^\prime$. Físicamente, no hay cambio en la energía y por tanto tampoco hay dispersión.
- $\theta = \pi$: el fotón retrocede. $\lambda^\prime =  \lambda + \frac{2h}{mc} $. El cambio en la longitud de onda (corrimiento) es máximo.
- $\theta = \pi$, y $mc^2 << E_\gamma$:
